### Question 1: Take 20 statements and generate the list of POS for each statement in the form of list or data frame.

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag

# Download the NLTK models
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
df1 = df.head(20)
```

```python
df1
```

```python
def get_pos_tags(statement):
    words = word_tokenize(statement)
    pos_tags = pos_tag(words)
    return [tag[1] for tag in pos_tags]
```

```python
pos_tags_list = [get_pos_tags(statement) for statement in df1['Statements']]
```

```python
for i, (statement, pos_tags) in enumerate(zip(df['Statements'], pos_tags_list), 1):
    print(f"{i}. Sentence: {statement.strip()}")
    for word, pos_tag in zip(word_tokenize(statement), pos_tags):
        print(f"   {word}: {pos_tag}")
    print()
```

**Code Components:**
1. Importing Necessary Libraries:
This section imports the required modules from the NLTK library for tokenization and POS tagging.

2. Downloading NLTK Models:
Downloads essential NLTK models for tokenization (punkt) and POS tagging (averaged_perceptron_tagger).

3. Subset DataFrame:
Creates a subset (df1) of the original DataFrame (df) containing the first 20 rows.

4. POS Tagging Function:
Defines a function (get_pos_tags) that takes a textual statement as input, tokenizes it into words, performs POS tagging, and returns a list of POS tags for each word.

5. Applying POS Tagging to DataFrame:
Creates a list (pos_tags_list) containing POS tags for each word in the 'Statements' column of the subset DataFrame.

6. Displaying Results:
Iterates through each statement in the DataFrame, printing the sentence followed by each word and its corresponding POS tag.

### Question 2: Generate the tokens from the 20 - 30 statements. Generate the word vectors of each token and list the similar words comparing the vectors

```python
import nltk
from nltk.tokenize import word_tokenize
from gensim.models import Word2Vec
```

```python
df2 = df.head(30)
```

```python
df2
```

```python
tokens = [word_tokenize(statement) for statement in df2['Statements']]
```

```python
model = Word2Vec(tokens, min_count=1)
word_vectors = {word: model.wv[word] for word in model.wv.index_to_key}
```

```python
for word in word_vectors:
    similar_words = model.wv.most_similar(word, topn=3)
    print(f"Similar words for '{word}': {', '.join(sim_word[0] for sim_word in similar_words)}")
```

**Code Components:**
1. Subset DataFrame:
The code starts by importing the necessary libraries and extracting a subset (df2) of the original DataFrame (df) containing the first 30 rows.

2. Tokenization:
The word_tokenize function from NLTK is used to tokenize each statement in the 'Statements' column, resulting in a list of tokenized statements (tokens).

3. Word2Vec Model Training:
A Word2Vec model is trained on the tokenized statements using Gensim's Word2Vec class. The min_count parameter sets the minimum number of occurrences for a word to be considered during training.

4. Similarity Analysis:
The trained Word2Vec model is used to obtain word vectors, and then, for each word, the three most similar words are identified and printed.
