### 1. Remove prefix, suffix, Exception, stop wards, numbers, symbols, URLs
### 2. Make sure the tokens are clean.

```python
import spacy
import re

# Load the English language model
nlp = spacy.load("en_core_web_sm")

def clean_text(text):
    # Remove URLs using regular expressions
    text = re.sub(r'http\S+', '', text)
    
    # Tokenize and process the text using spaCy
    doc = nlp(text)
    
    # Filter out tokens based on your criteria
    clean_tokens = []
    for token in doc:
        if not token.is_stop and not token.is_digit and not token.is_punct:
            # Remove prefixes and suffixes (lemmatization)
            clean_tokens.append(token.lemma_)
    
    return clean_tokens
```

```python
clean_tokens = clean_text(content)
```

### 3. Print the Text, POS, DEP, Lemmas of the tokens retrieved.

```python
for token in nlp(" ".join(clean_tokens)):
    print(f"Token: {token.text}, POS: {token.pos_}, DEP: {token.dep_}, Lemma: {token.lemma_}")
```
