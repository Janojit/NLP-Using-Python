### Removing prefix, suffix, Exception, stop wards, numbers, symbols, URLs
### Making sure the tokens are clean

```python
import spacy
import re

nlp = spacy.load("en_core_web_sm")

def clean_text(text):
    text = re.sub(r'http\S+', '', text)
    
    doc = nlp(text)
    
    clean_tokens = []
    for token in doc:
        if not token.is_space and not token.is_stop and not token.is_digit and not token.is_punct:
            clean_tokens.append(token.lemma_)
    
    return clean_tokens
```

```python
clean_tokens = clean_text(content)
```

### Printing the Text, POS, DEP, Lemmas of the tokens retrieved

```python
for token in nlp(" ".join(clean_tokens)):
    print(f"Token: {token.text}, POS: {token.pos_}, DEP: {token.dep_}, Lemma: {token.lemma_}")
```

## Question 1: Write a program to get synonyms from WordNet 

### Finding Synonyms of Each Tokens

#### Define a function get_synonyms_and_definitions(token)

```python
from nltk.corpus import wordnet as wn

def get_synonyms_and_definitions(token):
    synonyms_and_definitions = {}
    for synset in wn.synsets(token):
        synonyms = synset.lemma_names()
        definitions = synset.definition()
        for synonym in synonyms:
            synonyms_and_definitions[synonym] = definitions
    return synonyms_and_definitions
```

**Above Code Explaination:** 
- This function takes a token as input, which represents a word.
- It initializes an empty dictionary synonyms_and_definitions to store synonyms and their definitions.
- It iterates through the synsets (groups of synonymous words) in WordNet for the given token.
- For each synset, it extracts synonyms using synset.lemma_names() and the definition of the synset using synset.definition().
- It then associates each synonym with the definition in the synonyms_and_definitions dictionary.
- Finally, it returns the synonyms_and_definitions dictionary.

#### Iterate through the clean_tokens and store the synonyms and definition into a dictionary

```python
token_synonyms_definitions = {}

for token in clean_tokens:
    synonyms_and_definitions = get_synonyms_and_definitions(token)
    token_synonyms_definitions[token] = synonyms_and_definitions
```

**Above Code Explaination:** 
- Initialize an empty dictionary "token_synonyms_definitions"
- Then This loop goes through each token in the clean_tokens list, which represents the lemmatized and cleaned tokens from your input text.
- For each token, it calls the get_synonyms_and_definitions function to find synonyms and their definitions.
- It associates the token with the obtained synonyms and their definitions in the token_synonyms_definitions dictionary.

#### Print the results

```python
for token, synonyms_definitions in token_synonyms_definitions.items():
    print(f"Token: {token}")
    for synonym, definition in synonyms_definitions.items():
        print(f"  Synonym: {synonym}")
        print(f"    Definition: {definition}")
```

**Above Code Explaination:** 
- This loop iterates through the token_synonyms_definitions dictionary.
- For each token, it prints the token itself.
- Then, for each synonym associated with that token, it prints the synonym and its definition.

## Question 2: Write a program to get Antonyms from WordNet

### Finding Antonyms of Each Tokens

#### Define a function get_antonyms_with_definitions(token)

```python
def get_antonyms_with_definitions(token):
    antonyms_with_definitions = {}
    for synset in wn.synsets(token):
        for lemma in synset.lemmas():
            antonyms = [antonym.name() for antonym in lemma.antonyms()]
            for antonym in antonyms:
                ant_synsets = wn.synsets(antonym)
                if ant_synsets:
                    ant_synset = ant_synsets[0]  # Get the first sense (synset) of the antonym
                    antonyms_with_definitions[antonym] = ant_synset.definition()
    return antonyms_with_definitions
```

**Above Code Explaination:** 
- This function takes a token (a word) as input.
- It initializes an empty dictionary antonyms_with_definitions to store antonyms and their definitions.
- It iterates through the synsets (groups of synonymous words) in WordNet for the given token.
- For each synset, it iterates through the lemmas and retrieves antonyms using lemma.antonyms().
- For each antonym, it checks if there are synsets associated with it using wn.synsets(antonym).
- If there are synsets for the antonym, it gets the first sense (synset) of the antonym and associates it with its definition in the antonyms_with_definitions dictionary.
- Finally, it returns the antonyms_with_definitions dictionary.

#### Iterate through the clean_tokens and store the antonym and definition into a dictionary

```python
token_antonyms_definitions = {}

for token in clean_tokens:
    antonyms_with_definitions = get_antonyms_with_definitions(token)
    token_antonyms_definitions[token] = antonyms_with_definitions
```

**Above Code Explaination:** 
- Initialize an empty dictionary "token_antonyms_definitions"
- This loop goes through each token in the clean_tokens list, representing lemmatized and cleaned tokens from your input text.
- For each token, it calls the get_antonyms_with_definitions function to find antonyms and their definitions.
- It associates the token with the obtained antonyms and their definitions in the token_antonyms_definitions dictionary.

#### Print the results

```python
for token, antonyms_definitions in token_antonyms_definitions.items():
    print(f"Token: {token}")
    for antonym, definition in antonyms_definitions.items():
        print(f"  Antonym: {antonym}")
        print(f"    Definition: {definition}")
```

**Above Code Explaination:** 
- This loop iterates through the token_antonyms_definitions dictionary.
- For each token, it prints the token itself.
- Then, for each antonym associated with that token, it prints the antonym and its definition.
