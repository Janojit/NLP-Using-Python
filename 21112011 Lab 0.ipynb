## Question 1: Read the file and split the words

### Reading the File

```python
with open('Article of Football.txt','r',encoding='utf-8') as f:
    content = f.read()
```

```python
print(content)
```

```python
content = content.strip()
```

```python
print(content)
```

### Spliting the Text and Converting into List

```python
line = content.split(".")
```

```python
words = []
for i in range(len(line)):
    words+=line[i].split()
```

```python
words
```

```python
for i in range(len(words)):
    words[i] = words[i].lower()
```

## Question 2: Remove the stop words

```python
print(f"Number of Words before Stopwords {len(words)}")
```

```python
stopwords = [
    'a', 'an', 'and', 'are', 'as', 'at', 'be', 'by', 'for', 'from','their','but',
    'has', 'he', 'in', 'is', 'it', 'its', 'of', 'on', 'that', 'the','they','which',
    'to', 'was', 'were', 'will', 'with', 'you','there','also','this','can','who',
    '.', ',', ';', ':', '!', '?', '-', '--', '(', ')', '[', ']', '{', '}', "'", '"'
]
```

```python
for i in range(len(stopwords)):
    if stopwords[i] in words:
        words = [x for x in words if x != stopwords[i]]
```

```python
words
```

```python
print(f"Number of Words after Stopwords {len(words)}")
```

## Question 3: Create an dictionary with the word count

```python
frequency = {}

for item in words:
    if item in frequency:
        frequency[item] += 1
    else:
        frequency[item] = 1

for key, value in frequency.items():
    print(f"{key} : {value}")
```

### Sorting in Descending Order

```python
sorted_frequency = dict(sorted(frequency.items(), key = lambda x: x[1], reverse = True))
```

```python
sorted_frequency
```

## Question 4: Plot the graph to visualize the extracted words

### Selecting The Top 25 Occuring Words

```python
sorted25 = dict(list(sorted_frequency.items())[0: 25])
```

```python
sorted25
```

### Histogram

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(25,15))
plt.bar(sorted25.keys(), sorted25.values())
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Word Frequency')
plt.xticks(rotation=90)
plt.show()
```

### Pie Plot

```python
wor = []
count = []

for key, value in sorted25.items():
    wor.append(key)
    count.append(value)

plt.pie(count, labels=wor)

plt.axis('equal')
plt.show()
```
